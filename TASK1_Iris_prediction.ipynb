## **TASK 1 (BEGINNER LEVEL): Iris Flowers Classification**

This is perhaps the best-known example in the field of machine learning.The aim is to classify iris flowers among three species (Setosa, Versicolor, or Virginica) from sepals' and petals' length and width measurements. The iris data set contains fifty instances of each of the three species. The central goal here is to design a model that makes useful classifications for new flowers.

```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
col_name = ['SepalLength','SepalWidth','PetalLength','PetalWidth','Class']
```

```python
df=pd.read_csv(url,names=col_name)
```

## **Encoding**

```python
x=df.drop(["Class"],axis=1)
y=df[["Class"]]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
y_en=le.fit_transform(y)
```

```python
y_en
```

## **Training Model**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y_en,random_state=0,test_size=0.25)
```

## **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
model=lr.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
model.score(x,y_en)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

## **Evaluation**

```python
from sklearn.metrics import confusion_matrix
cnf=confusion_matrix(y_test,y_pred,labels=[0,1,2])
cnf
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
print(f'Accuracy of Model: {accuracy_score(y_test, y_pred)}')
print(f'Report of Model: \n {classification_report(y_test,y_pred)}')
```
